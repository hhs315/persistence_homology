```python
import numpy as np
import pandas as pd
from random import sample
import gudhi as gd
import random

import matplotlib.pyplot as plt
from sklearn import manifold

##n개의 점에 대해서 호몰로지를 계산하는 예시 입니다.
```

```python
matrix = np.zeros((10,10)) #10개의 점에 대해서 계산 하는 예시, 점 10개에 대한 서로간의 거리정보를 입력, 매트릭스는 대칭이 되어야함 왜냐하면 두 점간의 거리는 항상 같아야 하기때문.
for i in range(10):
    for j in range(i):
        matrix[i][j] = random.random()
        matrix[j][i] =matrix[i][j]

for i in range(10):
    matrix[i][i] = 0  #자기 자신과의 거리는 0이 되어야 한다.
```

```python
rips_complex = gd.RipsComplex(distance_matrix = matrix , max_edge_length = 1) #컴플렉스 구조를 만드는 명령어입니다. max_edge_length는 반지름을 어디까지 늘릴껀지 정합니다.
simplex_tree = rips_complex.create_simplex_tree(max_dimension=3) #위 구조로부터 트리를 만들어 각 생성점에 대한 생긴 시점과 사라진 시점에 생성점들을 합쳐가며 만듭니다.

diag_3 = simplex_tree.persistence() #트리로부터 다이어그램을 만듭니다.

gd.plot_persistence_diagram(diag_3) #다이어그램을 그려봅니다.
```
